```python
# agg method in pandas
# agg method helps to specify which aggregration method to use in particular column
# We have to pass a dict in this function
# agg method --> Passing dict
gn = movies.groupby("Genre")
gnn = gn.agg(
    {
        "Runtime":"mean",
        "No_of_Votes":"sum",
        "Gross":"sum",
        "IMDB_Rating":"mean"

    }
)
gnn.sort_values(["IMDB_Rating"],ascending=False)
```

```python
# Using agg function we can also pass a list
# agg --> passing a list
gn.agg(
    {
        "Runtime":['min','max','mean'],
        "No_of_Votes":['sum'],
        "IMDB_Rating":['min','max','mean','sum']
    }
)
```

```python
# Looping in group object
# finding the maximum ratted movie of each genre
# note in the new version the append is replaced by pd.concat([df,value],ignore_index<>)
empty_df = pd.DataFrame(columns=movies.columns)
for group,data in gn:
  empty_df = pd.concat([empty_df,data[data["IMDB_Rating"] == data["IMDB_Rating"].max()]],ignore_index=True)

empty_df
```

```python
# apply function in pandas
# We can pass the built in function through this

# split -- apply -- custom logic apply

# Q. Finding the number of movie that startwith A
def foo(group):
  return group["Series_Title"].str.startswith("A").sum()

```

```python
gn.apply(foo)
```

```python
def group_return(group):
  group["Group_Rank"]  = group["IMDB_Rating"].rank(ascending = False)
  return group
```

```python
gn.apply(group_return)
```
