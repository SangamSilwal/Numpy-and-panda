```python
sample_data = pd.DataFrame(
    {
        "1st_point":[0.8,0.5,0.7,6.7,8.8],
        "2nd_point":[9.9,7.7,0.44,5.5,8.8]
    }
)
sample_data
```

```python
def distance(row):
  p1 = row["1st_point"]
  p2 = row["2nd_point"]
  return (p1+p2)

sample_data['Distance'] = sample_data.apply(distance,axis=1)
```
