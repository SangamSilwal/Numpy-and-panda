```python
import pandas as pd
import numpy as np
```

```python
courses = pd.read_csv("/content/courses.csv")
dec = pd.read_csv("/content/reg-month1.csv")
nov = pd.read_csv("/content/reg-month2.csv")
students = pd.read_csv("/content/students.csv")
```

```python
# pd.concat
# ignore_index
# mulltiindex -> fetch using iloc
# concat dataframe horizontally
```

```python
reg = pd.concat([nov,dec],ignore_index = True)
```

```python
# multi index dataframe
multi = pd.concat([nov,dec],keys= ["nov","dec"])
multi.loc[('dec'),7]
```

```python
pd.concat([nov,dec],axis=1)   #this will horizontally stack the dta
```

```python
# merge
```

```python
# inner join
# basically in inner join the nan values are not consider
# the syntax is : leftdf.pd(rightdf,how="inner",on="column-name")
students.merge(nov,how="inner",on="student_id")
```

```python
# left
courses.merge(reg,how="left",on="course_id")
```

```python
# right
# outer
students.merge(reg,how="outer",on="student_id")
# outer will accept all the value though it is nan or not
```

```python
# finding the revenue made on each month
courses_detail = pd.concat([nov,dec],keys=['nov','dec']).reset_index()
courses_detail.merge(courses,on="course_id").groupby("level_0")['price'].sum()
```
